```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the datasets
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

downtime_col = 'Downtime Events'

# Define bins (0 to 4)
bins = np.arange(0, 6)
labels = [str(i) for i in bins]

# Count occurrences per bin
before_counts = before_df[downtime_col].value_counts().reindex(bins, fill_value=0).sort_index()
after_counts = after_df[downtime_col].value_counts().reindex(bins, fill_value=0).sort_index()

# Set up bar positions
x = np.arange(len(bins))  # [0, 1, 2, 3, 4, 5]
bar_width = 0.4

# Plot
plt.figure()
plt.bar(x - bar_width/2, before_counts, width=bar_width, label='Before', color='tomato')
plt.bar(x + bar_width/2, after_counts, width=bar_width, label='After', color='mediumseagreen')

plt.xticks(x, labels)
plt.xlabel("Number of Downtime Events")
plt.ylabel("Frequency")
plt.title("Histogram: Downtime Events Distribution")
plt.legend()
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSVs
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Calculate values
voltage_mode_before = before_df['Voltage (V)'].mode()[0]
voltage_mode_after = after_df['Voltage (V)'].mode()[0]

thd_mode_before = before_df['THD (%)'].mode()[0]
thd_mode_after = after_df['THD (%)'].mode()[0]

response_mode_before = before_df['Response Time (ms)'].mode()[0]
response_mode_after = after_df['Response Time (ms)'].mode()[0]

pf_mode_before = before_df['Power Factor'].mode()[0]
pf_mode_after = after_df['Power Factor'].mode()[0]

downtime_mean_before = before_df['Downtime Events'].mean()
downtime_mean_after = after_df['Downtime Events'].mean()

# Define categories
categories_primary = ['Voltage (V)', 'THD (%)', 'Response Time (ms)']
categories_secondary = ['Power Factor', 'Downtime']

# Extract values
before_primary = [voltage_mode_before, thd_mode_before, response_mode_before]
after_primary = [voltage_mode_after, thd_mode_after, response_mode_after]

before_secondary = [pf_mode_before, downtime_mean_before]
after_secondary = [pf_mode_after, downtime_mean_after]

# Bar positions
x_primary = np.arange(len(categories_primary))
x_secondary = np.arange(len(categories_secondary)) + len(x_primary) + 1  # space between groups

fig, ax1 = plt.subplots(figsize=(10, 6))

# Primary y-axis bars
bar_width = 0.35
ax1.bar(x_primary - bar_width/2, before_primary, bar_width, label='Before (Primary)', color='tomato')
ax1.bar(x_primary + bar_width/2, after_primary, bar_width, label='After (Primary)', color='seagreen')
ax1.set_ylabel('Voltage / THD / Response Time')
ax1.set_xticks(list(x_primary) + list(x_secondary))
ax1.set_xticklabels(categories_primary + categories_secondary, rotation=15)
ax1.grid(axis='y')

# Create secondary y-axis
ax2 = ax1.twinx()
ax2.bar(x_secondary - bar_width/2, before_secondary, bar_width, label='Before (Secondary)', color='tomato', hatch='//')
ax2.bar(x_secondary + bar_width/2, after_secondary, bar_width, label='After (Secondary)', color='seagreen', hatch='//')
ax2.set_ylabel('Power Factor / Downtime')

# Combine legends from both axes
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

plt.title("Mode Analysis with Average Downtime (Dual Y-Axis)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load CSVs
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Calculate values
voltage_mode_before = before_df['Voltage (V)'].mode()[0]
voltage_mode_after = after_df['Voltage (V)'].mode()[0]

thd_mode_before = before_df['THD (%)'].mode()[0]
thd_mode_after = after_df['THD (%)'].mode()[0]

response_mode_before = before_df['Response Time (ms)'].mode()[0]
response_mode_after = after_df['Response Time (ms)'].mode()[0]

# Power factor to %
pf_mode_before = before_df['Power Factor'].mode()[0] * 100
pf_mode_after = after_df['Power Factor'].mode()[0] * 100

# Downtime % (assuming 5 events max)
max_downtime = 5
downtime_mean_before = (before_df['Downtime Events'].mean() / max_downtime) * 100
downtime_mean_after = (after_df['Downtime Events'].mean() / max_downtime) * 100

# Define categories
categories_primary = ['Voltage (V)', 'Response Time (ms)']
categories_secondary = ['THD (%)', 'Power Factor (%)', 'Downtime (%)']

# Extract values
before_primary = [voltage_mode_before, response_mode_before]
after_primary = [voltage_mode_after, response_mode_after]

before_secondary = [thd_mode_before, pf_mode_before, downtime_mean_before]
after_secondary = [thd_mode_after, pf_mode_after, downtime_mean_after]

# Bar positions
x_primary = np.arange(len(categories_primary))
x_secondary = np.arange(len(categories_secondary)) + len(x_primary) + 1  # space between groups

fig, ax1 = plt.subplots(figsize=(10, 6))

bar_width = 0.35

# Primary y-axis bars (Voltage and Response Time)
ax1.bar(x_primary - bar_width/2, before_primary, bar_width, label='Before (Primary)', color='tomato')
ax1.bar(x_primary + bar_width/2, after_primary, bar_width, label='After (Primary)', color='seagreen')
ax1.set_ylabel('Voltage (V) / Response Time (ms)')
ax1.set_xticks(list(x_primary) + list(x_secondary))
ax1.set_xticklabels(categories_primary + categories_secondary, rotation=15)
ax1.grid(axis='y')

# Secondary y-axis bars (THD, PF, Downtime %)
ax2 = ax1.twinx()
ax2.bar(x_secondary - bar_width/2, before_secondary, bar_width, label='Before (%)', color='tomato', hatch='//')
ax2.bar(x_secondary + bar_width/2, after_secondary, bar_width, label='After (%)', color='seagreen', hatch='//')
ax2.set_ylabel('THD / Power Factor / Downtime (%)')

# Combine both legends
handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(handles1 + handles2, labels1 + labels2, loc='upper left')

plt.title("Mode & Percentage Comparison (THD, PF, Downtime)")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Total downtime count
total_downtime_before = before_df['Downtime Events'].sum()
total_downtime_after = after_df['Downtime Events'].sum()

# Pie data
labels = ['Before Stabilization', 'After Stabilization']
sizes = [total_downtime_before, total_downtime_after]
colors = ['#FF6B6B', '#4ECDC4']  # coral red and aqua green
explode = (0.05, 0.1)  # Emphasize both slices

# Plot
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sizes, labels=labels, colors=colors, explode=explode,
    autopct='%1.1f%%', shadow=True, startangle=140, textprops={'fontsize': 14}
)

# Title and center styling
plt.title("Share of Total Downtime Events", fontsize=16, weight='bold')
plt.gca().set_aspect('equal')  # Equal aspect ratio = perfect circle

# Optional: Add labels with counts
for i, a in enumerate(autotexts):
    a.set_text(f"{sizes[i]} events\n({a.get_text()})")
    a.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Calculate improvements
thd_improvement = before_df['THD (%)'].mode()[0] - after_df['THD (%)'].mode()[0]
response_improvement = before_df['Response Time (ms)'].mode()[0] - after_df['Response Time (ms)'].mode()[0]
pf_improvement = (after_df['Power Factor'].mode()[0] - before_df['Power Factor'].mode()[0]) * 100  # to %
downtime_improvement = before_df['Downtime Events'].mean() - after_df['Downtime Events'].mean()

# Prepare data
improvement_values = [thd_improvement, response_improvement, pf_improvement, downtime_improvement]
labels = ['THD Reduction (%)', 'Faster Response (ms)', 'Power Factor Gain (%)', 'Downtime Decrease (events)']
colors = ['#FF6B6B', '#FFD93D', '#6BCB77', '#4D96FF']  # red, yellow, green, blue
explode = (0.05, 0.05, 0.05, 0.05)

# Plot
plt.figure(figsize=(9, 9))
wedges, texts, autotexts = plt.pie(
    improvement_values, labels=labels, colors=colors, explode=explode,
    autopct='%1.1f%%', startangle=150, shadow=True, textprops={'fontsize': 13}
)

# Title and formatting
plt.title("Relative Contribution of Improvements After Stabilization", fontsize=16, weight='bold')
plt.gca().set_aspect('equal')  # Equal aspect for circle

# Add raw value next to percentage
for i, a in enumerate(autotexts):
    a.set_text(f"{improvement_values[i]:.2f} units\n({a.get_text()})")
    a.set_fontweight('bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Helper function to plot individual pie charts
def plot_pie(metric_name, before_value, after_value, units, colors, explode=(0.05, 0.05)):
    sizes = [before_value, after_value]
    labels = ['Before', 'After']

    plt.figure(figsize=(6.5, 6.5))
    wedges, texts, autotexts = plt.pie(
        sizes, labels=labels, autopct='%1.1f%%', explode=explode,
        startangle=140, shadow=True, colors=colors, textprops={'fontsize': 13}
    )
    plt.title(f"{metric_name} Comparison", fontsize=15, weight='bold')

    for i, a in enumerate(autotexts):
        a.set_text(f"{sizes[i]:.2f} {units}\n({a.get_text()})")
        a.set_fontweight('bold')

    plt.gca().set_aspect('equal')
    plt.tight_layout()
    plt.show()

# Colors
colors1 = ['#FF6B6B', '#4ECDC4']   # red, aqua green
colors2 = ['#FFD93D', '#6BCB77']   # yellow, green
colors3 = ['#4D96FF', '#FF6B6B']   # blue, red
colors4 = ['#9D4EDD', '#00BBF9']   # purple, cyan

# 1. THD (%)
thd_before = before_df['THD (%)'].mode()[0]
thd_after = after_df['THD (%)'].mode()[0]
plot_pie("Total Harmonic Distortion (THD %)", thd_before, thd_after, '%', colors1)

# 2. Power Factor (%)
pf_before = before_df['Power Factor'].mode()[0] * 100
pf_after = after_df['Power Factor'].mode()[0] * 100
plot_pie("Power Factor (%)", pf_before, pf_after, '%', colors2)

# 3. Response Time (ms)
response_before = before_df['Response Time (ms)'].mode()[0]
response_after = after_df['Response Time (ms)'].mode()[0]
plot_pie("Response Time (ms)", response_before, response_after, 'ms', colors3)

# 4. Downtime Events (Total)
downtime_before_total = before_df['Downtime Events'].sum()
downtime_after_total = after_df['Downtime Events'].sum()
plot_pie("Total Downtime Events", downtime_before_total, downtime_after_total, 'events', colors4)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
before_df = pd.read_csv('/content/factory_data_before.csv')
after_df = pd.read_csv('/content/factory_data_after.csv')

# Extract values
thd_before = before_df['THD (%)'].mode()[0]
thd_after = after_df['THD (%)'].mode()[0]

pf_before = before_df['Power Factor'].mode()[0] * 100
pf_after = after_df['Power Factor'].mode()[0] * 100

response_before = before_df['Response Time (ms)'].mode()[0]
response_after = after_df['Response Time (ms)'].mode()[0]

downtime_before_total = before_df['Downtime Events'].sum()
downtime_after_total = after_df['Downtime Events'].sum()

# Pie chart setup
metrics = [
    ("THD (%)", thd_before, thd_after, '%', ['#FF6B6B', '#4ECDC4']),
    ("Power Factor (%)", pf_before, pf_after, '%', ['#FFD93D', '#6BCB77']),
    ("Response Time (ms)", response_before, response_after, 'ms', ['#4D96FF', '#FF6B6B']),
    ("Downtime Events", downtime_before_total, downtime_after_total, 'events', ['#9D4EDD', '#00BBF9']),
]

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, (title, before_val, after_val, unit, colors) in enumerate(metrics):
    sizes = [before_val, after_val]
    labels = ['Before', 'After']
    wedges, texts, autotexts = axes[i].pie(
        sizes, labels=labels, autopct='%1.1f%%', startangle=140,
        explode=(0.05, 0.05), colors=colors, shadow=True, textprops={'fontsize': 12}
    )
    axes[i].set_title(f"{title} Comparison", fontsize=14, weight='bold')
    axes[i].axis('equal')

    for j, a in enumerate(autotexts):
        a.set_text(f"{sizes[j]:.2f} {unit}\n({a.get_text()})")
        a.set_fontweight('bold')

plt.suptitle("Before vs After Stabilization — Metric Comparisons", fontsize=16, weight='bold')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
